### Этап 2. Разметка датасета

```python
import re
import os
import cv2
from ultralytics import YOLO
from pathlib import Path

model = YOLO("yolov8x-world.pt")
model.set_classes(["cartoon character. boy teenager. brown hair. large round white eyes."])

image_paths = [
    str(p) for p in Path("./dataset").glob("morty_*.jpg")
    if re.fullmatch(r"morty_\d+\.jpg", p.name)
]

for img_path in image_paths:
    results = model.predict(
        source=img_path,
        conf=0,
        verbose=False
    )

    r = results[0]
    img = cv2.imread(img_path)
    h, w = img.shape[:2]

    label_lines = []

    if r.boxes is not None:
        for box in [r.boxes[0]]:
            x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
            cx = ((x1 + x2) / 2) / w
            cy = ((y1 + y2) / 2) / h
            bw = (x2 - x1) / w
            bh = (y2 - y1) / h

            label_lines.append(f"0 {cx:.6f} {cy:.6f} {bw:.6f} {bh:.6f}")

            cv2.rectangle(
                img,
                (int(x1), int(y1)),
                (int(x2), int(y2)),
                (0, 255, 0),
                2
            )
            cv2.putText(
                img,
                "Rick and Morty: Morty",
                (int(x1), int(y1) - 5),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.6,
                (0, 255, 0),
                2
            )

    out_img_path = img_path.replace(".jpg", "_det.jpg")
    cv2.imwrite(out_img_path, img)

    out_txt_path = img_path.replace(".jpg", ".txt")
    with open(out_txt_path, "w") as f:
        f.write("\n".join(label_lines))

    print(f"[OK] {os.path.basename(img_path)}")

print("Авторазметка завершена.")
```

```python
import cv2
import os
import numpy as np

folder_path = "dataset"
output_folder = "collages"
cols, rows = 4, 4

os.makedirs(output_folder, exist_ok=True)

all_images = sorted([f for f in os.listdir(folder_path) 
                    if f.endswith('_det.jpg')],
                    key=lambda x: int(x.split('_')[1]))

images_per_page = cols * rows

for page in range(0, len(all_images), images_per_page):
    collage = np.ones((rows * 540, cols * 960, 3), dtype=np.uint8) * 255
    
    for i in range(images_per_page):
        idx = page + i
        if idx >= len(all_images):
            break
        
        img_path = os.path.join(folder_path, all_images[idx])
        img = cv2.imread(img_path)
        if img is not None:
            img_resized = cv2.resize(img, (960, 540))
            
            n = all_images[idx].split('_')[1]
            cv2.putText(img_resized, f"n={n}", (20, 40), 
                       cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

            row = i // cols
            col = i % cols
            y_start = row * 540
            x_start = col * 960
            
            collage[y_start:y_start+540, x_start:x_start+960] = img_resized
    
    collage_filename = f"collage_{page//images_per_page + 1:03d}.jpg"
    cv2.imwrite(os.path.join(output_folder, collage_filename), collage)
    print(f"Сохранен: {collage_filename}")
```

Окончательно подготовим данные для обучения

```python
import random, re
from pathlib import Path
import shutil

source_dir = 'dataset'
target_dir = 'data'
train_ratio = 0.8

train_images_dir = Path(target_dir) / 'images' / 'train'
val_images_dir = Path(target_dir) / 'images' / 'val'
train_labels_dir = Path(target_dir) / 'labels' / 'train'
val_labels_dir = Path(target_dir) / 'labels' / 'val'

for d in [train_images_dir, val_images_dir, train_labels_dir, val_labels_dir]:
    d.mkdir(parents=True, exist_ok=True)

image_files = [p for p in Path("./dataset").glob("morty_*.jpg")
    if re.fullmatch(r"morty_\d+\.jpg", p.name)]

random.shuffle(image_files)
split_idx = int(len(image_files) * train_ratio)
train_files = image_files[:split_idx]
val_files = image_files[split_idx:]

for img in train_files:
    shutil.copy2(img, train_images_dir / img.name)
    txt_path = img.with_suffix('.txt')
    if txt_path.exists():
        shutil.copy2(txt_path, train_labels_dir / txt_path.name)

for img in val_files:
    shutil.copy2(img, val_images_dir / img.name)
    txt_path = img.with_suffix('.txt')
    if txt_path.exists():
        shutil.copy2(txt_path, val_labels_dir / txt_path.name)
```

И непосредственно дообучим модель на 75 эпохах. Тут в качестве инструмента для трекинга экспериментов используется WandB.

```python
from ultralytics import YOLO
import os, wandb

wandb.init(
    project="morty-detection",
    name="yolov8-training",
    config={
        "model": "yolov8m",
        "dataset": "morty_custom",
        "epochs": 75,
        "batch": 16
    }
)

def log(trainer):
    wandb.log({
                "epoch": trainer.epoch,
                "train/box_loss": trainer.metrics.get('train/box_loss', 0),
                "train/cls_loss": trainer.metrics.get('train/cls_loss', 0),
                "train/dfl_loss": trainer.metrics.get('train/dfl_loss', 0),
                "val/box_loss": trainer.metrics.get('val/box_loss', 0),
                "val/cls_loss": trainer.metrics.get('val/cls_loss', 0),
                "val/dfl_loss": trainer.metrics.get('val/dfl_loss', 0),
                "precision": trainer.metrics.get('metrics/precision(B)', 0),
                "recall": trainer.metrics.get('metrics/recall(B)', 0),
                "mAP50": trainer.metrics.get('metrics/mAP50(B)', 0),
                "mAP50-95": trainer.metrics.get('metrics/mAP50-95(B)', 0)
            })

model = YOLO("yolov8m.pt")
model.add_callback("on_fit_epoch_end", log)

os.chdir("./data/")
model.train(data="./data.yaml", batch=16, epochs=75,
            save=True, exist_ok=True, project='morty_project',
    name='train_v1', save_period=1, plots=True, cache=True, save_json=True)

wandb.finish()
```

```python
import os
import cv2
import numpy as np
from pathlib import Path

dataset_path = "dataset"
output_path = "results_pairs"

os.makedirs(output_path, exist_ok=True)

all_n = set()
for file in Path(dataset_path).glob("morty_*.jpg"):
    parts = file.stem.split('_')
    if len(parts) >= 2:
        all_n.add(parts[1])

for n in sorted(all_n, key=int):
    det_path = Path(dataset_path) / f"morty_{n}_det.jpg"
    pred_path = Path(dataset_path) / f"morty_{n}_pred.jpg"

    det_img = cv2.imread(str(det_path))
    pred_img = cv2.imread(str(pred_path))

    cv2.putText(det_img, "det", (20, 50), 
           cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255, 0), 3)
    cv2.putText(pred_img, "pred", (20, 50),
           cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 3)
    
    combined = np.vstack([det_img, pred_img])
    
    cv2.imwrite(os.path.join(output_path, f"morty_{n}_pair.jpg"), combined)
    print(f"✅ n={n}: {det_img.shape} + {pred_img.shape}")
```
